# Verosimilitud de una muestra

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

## Datos

```python
datos = np.array([0.32, 1.02, 5.07, 1.28, 1.61])
```

```python
ndatos = len(datos)
ndatos
```

```python
tmedio = np.mean(datos)
tmedio
```

![](https://drive.google.com/uc?id=15vY0LKY9TsTS85H1t3H2KiyU7Y9I0y47)

## Verosimilitud

```python
tau_max = tmedio * 5
tau = np.linspace(0.01, tau_max, 256)
```

```python
verosimilitud = np.power( np.exp(-tmedio/tau)/tau, ndatos)
```

## Graficar

```python
fig, ax = plt.subplots()
ax.set_xlabel(r'$\tau$')
ax.set_ylabel(r'L($\tau$)')
ax.plot(tau, verosimilitud, label=r'L($\tau$)')
ax.axvline(tmedio, ls='--', color='tab:orange')
(ymin, ymax) = ax.get_ylim()
delta_y = ymax - ymin
ytext = ymin - 0.1 * delta_y
ax.text(tmedio, ytext, r'$\hat{\tau}$', color='tab:orange', horizontalalignment='center', verticalalignment='top')
plt.tight_layout()
plt.show()
```
