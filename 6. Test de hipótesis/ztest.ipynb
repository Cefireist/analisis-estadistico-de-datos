# Análisis Estadístico de Datos


## Test del aditivo para pintura

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

Significancia del test

```python
alpha = 0.05
```

Z crítico (cola inferior)

```python
z_critico = stats.norm.ppf(alpha)
print(f'z crítico = {z_critico:.2f}')
```

Estadística Z

```python
mu_x = 75
sigma = 9
```

```python
n = 25
xobs = 70.8
```

```python
sigma_media = sigma / math.sqrt(n)
sigma_media
```

```python
zobs = (xobs - mu_x)/sigma_media
zobs
```

Gráfico

```python
fig = plt.figure(figsize=(16,6))
ax = fig.subplots()
fsize, lsize = 18, 12 # tamaño de fuentes
ax.set_xlabel('z')
ax.set_ylabel('f(z)', fontsize=fsize)
xmax = 3
z = np.linspace(-xmax, xmax, 256)
ax.plot(z, stats.norm.pdf(z))
ax.axvline(0, ls='--')
ax.tick_params(axis='both', which='major', labelsize=lsize)
z = np.linspace(-xmax, z_critico, 256)
ax.fill_between(z, 0, stats.norm.pdf(z), alpha=0.5)
ax.axvline(zobs, ls='--', color='tab:orange')
ax.text(zobs, 0.5, ' $z_{obs}$',
        transform=ax.get_xaxis_text1_transform(0)[0], fontsize=fsize)
plt.tight_layout()
plt.show()
```
