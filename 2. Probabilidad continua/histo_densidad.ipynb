# Análisis Estadístico de Datos

# Histogramas de frecuencia y densidad

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## Simulación de los datos

```python
rng = np.random.default_rng()
numero_eventos = 1000
datos = rng.normal(size=numero_eventos)
```

```python
datos
```

## Histograma de frecuencia

```python
xmin, xmax = -5, 5
histo_frecuencias, borde_bin = np.histogram(datos, bins=20, range=[xmin, xmax])
centro_bin = ( borde_bin[0:-1] + borde_bin[1:] ) / 2
```

```python
histo_frecuencias
```

```python
borde_bin
```

```python
centro_bin
```

Factor de escala: $ n \, \Delta x $

```python
ancho_bin = borde_bin[1] - borde_bin[0]
factor_escala = numero_eventos * ancho_bin
print("Factor de escala: ", factor_escala)
```

```python
fig, ax = plt.subplots()
ax.set_xlabel('X')
ax.set_ylabel('Frecuencia')

ax.plot(centro_bin, histo_frecuencias, ls='None', marker='o', color='tab:blue', label='$k_i$')
x = np.linspace(xmin, xmax, 256)
ax.plot(x, norm.pdf(x)*factor_escala, color='tab:orange', label='PDF')
ax.legend()
plt.show()
```

## Histograma de densidad


```python
fig, ax = plt.subplots()
ax.set_xlabel('X')
ax.set_ylabel('Densidad')

histo_densidad = histo_frecuencias / factor_escala
ax.plot(centro_bin, histo_densidad, ls='None', marker='o', color='tab:blue', label='$y_i$')
ax.plot(x, norm.pdf(x), color='tab:orange', label='PDF')
ax.legend()
plt.show()
```
